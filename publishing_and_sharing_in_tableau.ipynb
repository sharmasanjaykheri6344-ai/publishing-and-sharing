
Q.1 what is the key difference between publishing a dashboard on Tableau Public versus Tableau Server/
Cloud in terms of accessibility and data security?

...>> ✅ 1. Accessibility
🌐 Tableau Public
Dashboards are publicly accessible to anyone on the internet
No login required to view
Can be shared via link or embedded in websites
Suitable for:
Portfolio projects
Academic assignments
Public storytelling dashboards

👉 Example: If you publish sales analysis here, anyone worldwide can open and download it.

🏢 Tableau Server / ☁️ Tableau Cloud
Dashboards are private and access-controlled
Only authorized users (employees, teams, clients) can view
Requires login credentials
Supports:
Role-based access
Department-level sharing
Secure collaboration

👉 Example: A company can allow only regional managers to see their region’s sales dashboard.

🔐 2. Data Security
🚨 Tableau Public
No data security — data becomes publicly visible
Workbooks and underlying data can be downloaded by others
Not suitable for:
Confidential business data
Customer information
Financial or operational reports
🛡️ Tableau Server / Tableau Cloud
High data security and governance
Features include:
User authentication
Permissions & role-based access
Row-Level Security
Data encryption
Centralized data management

👉 Safe for real-world enterprise dashboards.

Tableau Public is best for open sharing and learning.

Q.2 When publishing a Tableau dashboard, what happens to a live connection compared to an extract?
How does this affect data refresh and performance?

....>> 🔗 Live Connection (Direct Connection to Source)
✅ What Happens When Published
The dashboard remains connected to the original data source (database, cloud warehouse, etc.).
Every time a user opens or interacts with the dashboard, queries are sent to the source system in real time.
🔄 Data Refresh
No manual refresh required — data is always up-to-date (real-time or near real-time).
Changes in the source database appear instantly in the dashboard.
⚡ Performance Impact
Performance depends on the data source speed and network latency.
May be slower if:
Database is large
Many users are querying simultaneously
Internet / server performance is weak

👉 Best for: Operational dashboards, real-time monitoring, frequently changing data.

📦 Extract (Snapshot of Data Stored in Tableau)
✅ What Happens When Published
Tableau publishes a compressed snapshot (.hyper extract) of the data.
The dashboard queries the extract stored on Tableau Server or Tableau Cloud instead of the original database.
🔄 Data Refresh
Data is not automatically real-time.
Requires:
Scheduled refresh
Manual refresh
Incremental refresh

👉 Until refresh happens → dashboard shows old (stale) data.

⚡ Performance Impact
Usually much faster because:
Data is optimized & compressed
Queries run on Tableau’s engine
Reduced load on the source database

👉 Best for: Large datasets, historical analysis, high-user dashboards.

⭐ Quick Comparison
Feature	Live Connection 🔗	Extract 📦
Data Freshness	Real-time	Depends on refresh schedule
Refresh Needed	❌ No	✅ Yes
Performance	Depends on source	Faster (optimized)
Load on Database	High	Low
Offline Availability	❌ No	✅ Yes (cached extract)
🎯 Final Concept (Exam Ready)
Live connection = real-time data but performance depends on the database.
Extract = faster performance but requires scheduled refresh to stay updated.

Q.3 In Tableau Server or Cloud (conceptually), which feature allows administrators to control which users
can view, interact with, or edit a published dashboard?

...>> In Tableau Server or Tableau Cloud, the main feature that allows administrators to control who can view, interact with, or edit a published dashboard is called 🔐 Permissions (Access Control / Role-Based Permissions).

🔐 Permissions in Tableau (Concept)

Permissions define what actions each user or group can perform on published content such as dashboards, workbooks, or data sources.

✅ Administrators can control:
👀 View → Who can open and see the dashboard
🔎 Interact → Who can filter, drill down, download images/data
✏️ Edit → Who can modify or republish the workbook
⬇️ Download → Who can download the workbook or data
🗑️ Delete / Publish → Who can manage content
👥 How It Is Managed

Permissions are typically assigned based on:

⭐ User Roles (Site Roles)

Examples:

Viewer
Explorer
Creator

These define the overall capability level of a user on the platform.

⭐ Project / Workbook Permissions

Admins can set:

Permissions at Project level (applies to all dashboards inside)
Permissions at Workbook or Dashboard level (more granular control)
🧠 Important Related Concept (Advanced Security)
Row-Level Security (RLS) can further restrict which data inside the dashboard a user can see, even if they can open the dashboard.
🎯 Exam-Ready One Line Answer

👉 In Tableau Server/Cloud, administrators use Permissions (Role-Based Access Control) to control which users can view, interact with, or edit a published dashboard.

Q.4 If you want users to view a dashboard but not download the underlying data, which permission setting
would you adjust in Tableau Server?

...>> In Tableau Server, if you want users to view a dashboard but NOT download the underlying data, you need to adjust the 🔐 “Download Full Data” (or “Download Data”) permission.

✅ Permission to Adjust

👉 Set “Download Full Data” = Deny (or remove this capability)

This allows users to:

👀 View the dashboard
🔎 Interact with filters / drill-downs
📊 Explore visuals

But prevents them from:

⬇️ Downloading the underlying data (raw data behind charts)
Exporting detailed datasets
🧠 Important Note

There are usually multiple download-related permissions. For strong data protection, administrators often control:

Download Full Data → Prevents access to raw underlying data ✅ (Most important here)
Download Summary Data → Optional (can allow or deny)
Download Workbook / Save As → Prevents copying the dashboard
🎯 Exam-Ready One Line Answer

👉 To allow viewing but prevent downloading underlying data in Tableau Server, administrators should deny the “Download Full Data” permission.

Q.6 List two best practices you should follow when preparing a dashboard for public sharing to ensure data
security and privacy?

...>> ✅ 1. Remove or Anonymize Sensitive Data

Before publishing:

❌ Do NOT include:
Customer names
Email IDs / phone numbers
Financial details
Confidential business metrics
✅ Instead:
Use aggregated data (e.g., total sales by region)
Replace real identifiers with dummy or anonymized values
Hide or remove sensitive columns

👉 Public dashboards can be downloaded and viewed by anyone, so raw data exposure is a major risk.

✅ 2. Limit Data Access (Hide Underlying Data / Reduce Detail)
Avoid showing transaction-level or row-level data
Remove unnecessary fields from the data source
Disable downloads where possible (if using secure platforms like Tableau Server or Tableau Cloud)
Publish only what is required for storytelling or analysis

👉 The less detailed data you expose, the lower the privacy risk.

🎯 Exam-Ready One-Line Answer

👉 Two best practices are: (1) remove or anonymize sensitive data, and (2) limit access to underlying or detailed data before public sharing.

Q.7 After publishing a dashboard, users report a “Data not found” error.
 What is the most common cause related to data source settings?

 ...>> After publishing a dashboard on Tableau Server or Tableau Cloud, the most common reason users see a “Data not found” error is:

❗ Broken or Incorrect Data Source Connection

This usually happens when the published dashboard cannot access the original data source.

✅ Most Common Causes (Data Source Settings Related)
🔐 1. Missing / Incorrect Database Credentials
The workbook was published using local or personal credentials
Credentials were not embedded during publishing
Server cannot authenticate → data connection fails

👉 Fix: Edit data source → Embed credentials or set proper authentication.

📁 2. Local File Path Not Accessible
Dashboard was connected to:
Excel file
CSV
Local database
After publishing, Tableau Server/Cloud cannot access your computer’s local path

👉 Fix:

Upload the file as an extract
Or place the file on a shared network location
🌐 3. Network / Firewall / Gateway Issue
Database is inside a private network
Server/Cloud cannot reach it

👉 Fix: Configure Tableau Bridge or network access.

🎯 Exam-Ready One Line Answer

👉 The most common cause of a “Data not found” error after publishing is that the data source connection is broken or credentials were not properly embedded, so the server cannot access the data.

Q.8 When embedding a Tableau dashboard into a secure internal portal, what two additional security
measures should be implemented to ensure only authorized users can access the embedded content?

...>> When embedding a dashboard from Tableau Server or Tableau Cloud into a secure internal portal, you should implement extra security controls to make sure only authorized users can access the embedded content 🔐.

Here are two key security measures ✅:

✅ 1. User Authentication (SSO / Trusted Authentication / OAuth)
Ensure users must log in before viewing the embedded dashboard
Common approaches:
Single Sign-On (SSO) integration
Trusted authentication (for internal apps)
OAuth / Identity provider login

👉 This verifies who the user is before Tableau allows access.

✅ 2. Permissions & Role-Based Access Control
Configure project/workbook permissions in Tableau
Assign users or groups capabilities such as:
View
Interact
Download
Edit

👉 This ensures users can only access dashboards they are authorized to see.

⭐ Bonus (Advanced but Important)
Implement Row-Level Security (RLS) so even authorized users see only their relevant data.
🎯 Exam-Ready One-Line Answer

👉 When embedding a Tableau dashboard in a secure portal, implement user authentication (such as SSO) and role-based permissions to ensure only authorized users can access the embedded content.
